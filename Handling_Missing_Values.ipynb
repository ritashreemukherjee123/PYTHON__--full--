**Methods of Identifying Missing data**

```python
df.isnull()   #Function which returns dataframe with only boolean values where True means "missing" and False means "not missing".
```

```python
df.notnull()   #opposite to the previous function
```

```python
df.info()   #Non-Null Count column gives no.of non-missing values for each column
```

```python
df.isna()    #Similar to .isnull() and .notnull() functions, only it shows True when the missing value is NaN type.
```

```python
df.duplicated()   #Identified the redundant row of data (here, 19 and 20 are duplicates)
```

**Handling Missing Values**

Rows 19 and 20 are duplicates(redundant values).

```python
#Removes the duplicates
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# Fill missing values in numeric columns with the mean of each column
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing values in non-numeric columns with 'Unknown'
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df[non_numeric_cols] = df[non_numeric_cols].fillna('Unknown')

display(df)
```

```python
import pandas as pd
from google.colab import files

# Upload the Excel file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the Excel file into a pandas DataFrame
try:
    df = pd.read_excel(filename)
    print(f"Successfully loaded {filename}:")
    display(df)
except Exception as e:
    print(f"Error reading the Excel file: {e}")
```

```python
#Removes the duplicates
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# Fill missing values in numeric columns with the mean of each column
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing values in non-numeric columns with 'Unknown'
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df[non_numeric_cols] = df[non_numeric_cols].fillna('Unknown')

display(df)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category in 'Paying Customer' and 'Do_Not_Contact' columns
paying_customer_counts = df['Paying Customer'].value_counts()
do_not_contact_counts = df['Do_Not_Contact'].value_counts()

# Create subplots for the two bar charts
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the 'Paying Customer' counts
axes[0].bar(paying_customer_counts.index, paying_customer_counts.values)
axes[0].set_title('Distribution of Paying Customers')
axes[0].set_xlabel('Paying Customer')
axes[0].set_ylabel('Count')

# Plot the 'Do_Not_Contact' counts
axes[1].bar(do_not_contact_counts.index, do_not_contact_counts.values, color='salmon')
axes[1].set_title('Distribution of Do Not Contact')
axes[1].set_xlabel('Do Not Contact')
axes[1].set_ylabel('Count')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```
